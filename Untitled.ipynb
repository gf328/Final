```python
url = "https://www.kaggle.com/datasets/nilimajauhari/amazon-prime-tv-shows"
html_page = urlopen(url)
html_text = html_page.read().decode("utf-8")
```

```python
for string in ["S.no.", "Age of viewers"]:
    string_start_idx = html_text.find(string)
    text_start_idx = string_start_idx + len(string)

    next_html_tag_offset = html_text[text_start_idx:].find("<")
    text_end_idx = text_start_idx + next_html_tag_offset

    raw_text = html_text[text_start_idx : text_end_idx]
    clean_text = raw_text.strip(" \r\n\t")
    print(clean_text)
```

```python
from urllib import request, error
from bs4 import BeautifulSoup

site = 'https://en.wikipedia.org/wiki/Gossip_Girl'

response = BeautifulSoup(request.urlopen(site),'html.parser')
TAB = response.find('table', class_='wikitable sortable')

print(TAB)


rows = TAB.find_all('tr')

with open('GossipGirl.csv', 'w') as file:
    file.write('Season, Timeslot, SeasonPremiere, SeasonFinale, TVSeason, Rank, USviewers, Rating')
    for row in rows:
        cells = row.find_all('td')
        row_contents = []
        for cell in cells:
            # Write the contents of the cell to the file
            row_contents.append(cell.text.strip())
        print(row_contents)    
        file.write(','.join(row_contents) + '\n')   
            

```
